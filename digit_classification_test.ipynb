```python
import numpy as np
import cv2
import pickle
```

```python
width = 600
height = 480
threshold = 0.75
```

```python
cap = cv2.VideoCapture(0)
cap.set(3, width)
cap.set(4, height)
```

```python
pickle_in = open("model_trained_20_epoch.p", "rb")
model = pickle.load(pickle_in)
```

```python
def preProcessing(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.equalizeHist(img)
#     img = img/255
    img = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
    return img
```

```python
while True:
    succes, imgOriginal = cap.read()
    img = np.asarray(imgOriginal)
    img = cv2.resize(img, (32, 32))
    img = preProcessing(img)
    cv2.imshow("Processed Image", img)
    img = img.reshape(1, 32, 32, 1) # reshape image before sending it for prediction
    
    # prediction
    classIndex = int(model.predict_classes(img)) # first predict the class ID
#     print(classIndex)
    predictions = model.predict(img)
    # returns the maximum value
    probVal = np.amax(predictions)
    print(classIndex, probVal)
    
    if probVal > threshold:
        cv2.putText(imgOriginal, "Predicted Value: " + str(classIndex),
                    (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 1)
        cv2.putText(imgOriginal, "&Probability: " + str(probVal),
                    (50, 100), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 1)
    
    cv2.imshow("Original Image", imgOriginal)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows() 
    
```
